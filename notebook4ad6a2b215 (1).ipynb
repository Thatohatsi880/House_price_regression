```python
#import libraries
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_absolute_error
from sklearn.model_selection import train_test_split

#Load the data and separate the target
iowa_file_path = '/kaggle/input/home-data-for-ml-course/train.csv'
home_data = pd.read_csv(iowa_file_path)
y = home_data.SalePrice
y
```

```python
#creating X
features = ['LotArea', 'YearBuilt', '1stFlrSF', '2ndFlrSF', 'FullBath', 'BedroomAbvGr', 'TotRmsAbvGrd']

# Select columns corresponding to features and preview the data
X = home_data[features]
X.head(20)
```

```python
#Split into validation and training data
train_X, train_y, test_X, test_y = train_test_split(X, y, random_state = 1)
train_X, train_y, test_X, test_y 
```

```python
random_forest_model = RandomForestRegressor(random_state = 1)
random_forest_model.fit(X, y)
random_forest_val_predictions = random_forest_model.predict(X)
random_forest_val_mae = mean_absolute_error(random_forest_val_predictions, y)

print('Validation MAE for Random Forest Model: {:,.0f}'.format(random_forest_val_mae))
```

```python
# A path to file I will be using for predictions
test_path = '/kaggle/input/home-data-for-ml-course/test.csv'

# read test data file using pandas
df = pd.read_csv(test_path)
df
```

```python
#create test_X which comesfrom test_data but includes only the columns used for prediction
# The list of columns is stored in a variable called features
random_forest_on_full_data = RandomForestRegressor()
random_forest_on_full_data.fit(X, y)
test_X = df[features]
test_X
```

```python
#making predictions to submit
test_predictions = random_forest_on_full_data.predict(test_X)

result = pd.DataFrame({'Id': df.Id, 'SalePrice': test_predictions})
result.to_csv("submission_csv")
result
```
